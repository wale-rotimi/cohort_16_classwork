# 1. Which columns represent engagement metrics?
The engagement metrics are columns that measure user interaction with posts.
These typically include:
- Likes
- Comments
- Shares
- Views

These columns indicate how users interact with the content on the platform.

# 2. Are there any missing values?
Yes, there are missing values in the dataset.
This can be observed using the `df.info()` output, where some columns have
non-null counts less than the total number of rows.

# 3. Which column has the highest average value?
From the summary statistics of numeric columns, the column with the highest
mean (average) value represents the metric with the highest average engagement.
This is most likely the **Views** column, as views generally have higher values
than likes, comments, or shares.

```python
import numpy as np
nums = np.array ([1, 3, 5, 7, 9])
print ("The Array:", nums)

```

```python
nums2  = np.array(nums+2)
print(nums2)
```

```python
num3 = np.array(nums2*3)
print(num3)
```

```python
total = np.sum(nums)
maximum = np.max(nums)
mean = np.mean(nums)
print("Sum", total)
print("Maximum", maximum)
print("Mean", mean)
```

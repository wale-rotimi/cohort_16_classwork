```python
##Task2
import numpy as np

# Create the NumPy array
visits = np.array([120, 135, 150, 160, 155, 170, 180, 175, 190, 200, 210, 205, 220, 230])

# 1. Check the type of the dataset
dataset_type = type(visits)

# 2. Find the total number of elements
total_elements = visits.size

# 3. Calculate the total number of visits (sum)
total_visits = visits.sum()

# Displaying the results
print(f"Dataset Type: {dataset_type}")
print(f"Total Number of Elements: {total_elements}")
print(f"Total Sum of Visits: {total_visits}")
```

```python
## Task3
# 1. Increase all daily visits by 15%
growth_simulated = visits * 1.15

# 2. Normalize the data using Min-Max scaling
# Formula: (x - min) / (max - min)
visits_min = visits.min()
visits_max = visits.max()
normalized_visits = (visits - visits_min) / (visits_max - visits_min)

# 3. Round the normalized values to 2 decimal places
normalized_rounded = np.round(normalized_visits, 2)

print(f"Growth Simulated: {growth_simulated}")
print(f"Normalized & Rounded: {normalized_rounded}")
```

```python
import numpy as np

# Create the NumPy array
nums = np.array([1, 3, 5, 7, 9])
print("Original array:", nums)
```

```python
# Add 2 to each element and store in a new array
nums_plus_2 = nums + 2
print("After adding 2:", nums_plus_2)
```

```python
# Multiply each element by 3
nums_times_3 = nums * 3
print("After multiplying by 3:", nums_times_3)
```

```python
# NumPy methods on the original array
print("Sum of original array:", np.sum(nums))
print("Maximum value:", np.max(nums))
print("Mean value:", np.mean(nums))
```
